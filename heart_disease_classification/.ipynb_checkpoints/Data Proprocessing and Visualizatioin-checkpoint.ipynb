```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_theme(style="darkgrid")
```

## 1. Data Preprocessing

```python
"""
age: The person's age in years
sex: The person's sex (1 = male, 0 = female)
cp: The chest pain experienced (Value 1: typical angina, Value 2: atypical angina, Value 3: non-anginal pain, Value 4: asymptomatic)
trestbps: The person's resting blood pressure (mm Hg on admission to the hospital)
chol: The person's cholesterol measurement in mg/dl
fbs: The person's fasting blood sugar (> 120 mg/dl, 1 = true; 0 = false)
restecg: Resting electrocardiographic measurement (0 = normal, 1 = having ST-T wave abnormality, 2 = showing probable or definite left ventricular hypertrophy by Estes' criteria)
thalach: The person's maximum heart rate achieved
exang: Exercise induced angina (1 = yes; 0 = no)
oldpeak: ST depression induced by exercise relative to rest ('ST' relates to positions on the ECG plot. See more here)
slope: the slope of the peak exercise ST segment (Value 1: upsloping, Value 2: flat, Value 3: downsloping)
ca: The number of major vessels (0-3)
thal: A blood disorder called thalassemia (3 = normal; 6 = fixed defect; 7 = reversable defect)
target: Heart disease (0 = no, 1 = yes)
"""
data = pd.read_csv("./heart.csv")
data.head(5)
```

```python
data.info()
```

```python
data = data.astype('float32')
data.head(5)
```

```python
data.isna().sum()
```

```python
data.describe()
```

```python
plt.figure(figsize=(40,15))
ax=sns.boxplot(data=data.iloc[:,0:-1], orient='h', palette='Set1')
ax.set_title("Box Plot Across All Attr")
plt.show()
```

```python
"""remove outlier of the data"""
q1 = data.quantile(0.25)
q3 = data.quantile(0.75)
iqr = q3 - q1
pd.DataFrame(iqr,columns=['IQR'])
```

```python
(data<q1-1.5*iqr) | (data>q3+1.5*iqr)
```

```python
data_filtered = data[~((data<q1-1.5*iqr) | (data>q3+1.5*iqr))].dropna()
data_filtered
```

### 2. Data Visualization

```python
fig = plt.figure(figsize=(30,8))
for index, col_name in enumerate(data_filtered.columns.values):
    ax_subplot=plt.subplot(2,7,index+1)
    ax = sns.histplot(data=data_filtered[col_name], kde=True, stat='density', color='green')
    ax.set_title(f'{col_name}')
    ax.set_xlabel('value')
fig.suptitle("Attribute PDF")
fig.tight_layout()
fig.subplots_adjust(top=0.9)
plt.show()
```

```python
continuous_attr=['age','trestbps','chol','thalach','oldpeak']
for index, attr_name in enumerate(continuous_attr):
    sns.displot(data=data_filtered, x=attr_name,col='target', kind='kde',color='green')
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=data_filtered, x='target')
plt.title('Target Count')
plt.show()
ratio = data_filtered[data_filtered['target']==0]['target'].count()/data_filtered[data_filtered['target']==1]['target'].count()
print(f'negative instance/positive instance: {round(ratio,3)}')
```

```python
categorical_attr = ['sex', 'cp','fbs','restecg','exang','slope','ca','thal']
fig=plt.figure(figsize=(30,8))
for index, attr_name in enumerate(categorical_attr):
    ax_sub=plt.subplot(2,4,index+1)
    ax = sns.countplot(data=data_filtered, x='target', hue=attr_name)
    ax.set_title(f"target V.S. {attr_name}")
    ax.set_xlabel("target value")
fig.suptitle("target V.S. categorical attributes")
fig.tight_layout()
fig.subplots_adjust(top=0.9)
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(data_filtered.corr(method='pearson'), annot=True, cmap='Greens')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.barplot(y=data_filtered.columns.values[0:-1], x=data.corr(method='pearson')['target'][0:-1])
plt.title("Correlation Target V.S. Other Attributes")
plt.show()
```

```python
"""drop fbs column as it only has one value-0 after removing the outliers and also the correlation value is 0"""
data_filtered = data_filtered.drop('fbs',axis=1)
```
